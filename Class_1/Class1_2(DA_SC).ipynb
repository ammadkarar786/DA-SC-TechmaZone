### A. What is a Variable?
*   **Definition:** A variable is a named container that stores a value.
*   **Analogy:** Think of a variable like a **Roll Number** in a school register.
    *   The Roll Number (Variable Name) doesn't change, but it points to specific student data (Value).
    *   When you call the roll number, the system retrieves the specific percentage or name associated with it.
*   **Key Takeaway:** Variables allow us to store data now and retrieve/use it later in the program.

### B. Variable Declaration & Assignment
*   **Syntax:** `variable_name = value`
*   **The Assignment Operator (`=`):**
    *   It does **not** mean "equal to" in the mathematical sense.

### D. Rules for Legal Variable Names
Python has strict rules for naming variables. If these rules are broken, the code will throw a `SyntaxError`.

| Rule | Description | Legal Example | Illegal Example | Reason for Illegality |
| :--- | :--- | :--- | :--- | :--- |
| **Start Character** | Must start with a letter (a-z, A-Z) or underscore (`_`). | `_temp`, `name` | `1st_place` | Cannot start with a number. |
| **Subsequent Characters** | Can contain letters, numbers, or underscores. | `user_id_1` | `user-id` | Hyphens are not allowed (interpreted as minus sign). |
| **Case Sensitivity** | `Age` and `age` are different variables. | `Total`, `total` | N/A | Python is case-sensitive. |
| **Reserved Words** | Cannot use Python keywords (e.g., `if`, `for`, `while`). | `my_if` | `if` | `if` is a reserved keyword. |

### E. Printing & Data Types
*   **Printing Multiple Variables:** Use commas inside the `print()` function to separate items.
*   **Checking Data Types:** Use the `type()` function.
    *   `int`: Whole numbers (e.g., `5`, `-10`)
    *   `float`: Decimal numbers (e.g., `5.0`, `3.14`)
    *   `str`: Text/Strings (e.g., `"Hello"`, `"2.3"`)

### F. Special Assignment Operators (Shorthand)
Instead of writing out the full equation, use compound operators to update a variable’s current value.

| Operator | Example | Equivalent To | Description |
| :--- | :--- | :--- | :--- |
| `+=` | `x += 10` | `x = x + 10` | Add and assign |
| `-=` | `x -= 5` | `x = x - 5` | Subtract and assign |
| `*=` | `x *= 2` | `x = x * 2` | Multiply and assign |
| `/=` | `x /= 4` | `x = x / 4` | Divide and assign |

### G. Type Casting
Type casting is converting a value from one data type to another.

1.  **Int to Float:** Adds a decimal point.
    *   `float(5)` → `5.0`
2.  **Float to Int:** Removes the decimal portion (truncates, does not round).
    *   `int(5.9)` → `5`
3.  **String to Number:** Only works if the string contains valid numeric characters.
    *   `int("10")` → `10`
    *   `float("2.3")` → `2.3`
4.  **Invalid Casting:** Alphabetical strings cannot become numbers.
    *   `int("Ali")` → **Error!**

## 4. Common Student Misconceptions to Address
1.  "=" means equals: Emphasize that `=` is an *assignment* operator, not a mathematical equality check.
2.  Variable Overwriting: Students often think a variable holds all its past values. Clarify that assigning a new value *replaces* the old one entirely.
3.  Casting Errors: Remind students that `int("Hello")` will crash the program. Only numeric strings can be cast to numbers.
4.  Float to Int Truncation: Students expect `int(5.9)` to be `6`. Explain that Python simply *cuts off* the decimal, it does not round up.
