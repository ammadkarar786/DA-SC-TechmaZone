## 1. Fundamental Data Types
*Goal: Understand what kind of data we are working with.*

| Data Type | Description | Examples |
| :--- | :--- | :--- |
| **Integer (`int`)** | Whole numbers without decimals. | `1`, `55`, `-10` |
| **Float (`float`)** | Numbers with decimal points. | `2.12`, `2.5`, `3.0` |
| **String (`str`)** | Sequence of characters enclosed in quotes (`' '` or `" "`). Can contain letters, numbers, symbols. | `"Hello World"`, `"daniyal98@gmail.com"` |

### Essential Built-in Functions
*   **`print()`**: Displays output to the screen.
    *   `print(2)` → Output: `2`
    *   `print("Hello World")` → Output: `Hello World`
*   **`type()`**: Reveals the data type class of a value.
    *   `type(2)` → Output: `<class 'int'>`
    *   `type(2.5)` → Output: `<class 'float'>`

> ⚠️ **Common Error Alert:** You cannot directly add an integer and a string (e.g., `5 + "hello"` will cause an error). They must be compatible types.

## 2. Arithmetic Operators
*Goal: Perform mathematical calculations in Python.*

| Operator | Name | Description | Example | Result |
| :---: | :--- | :--- | :--- | :---: |
| `+` | Addition | Adds values | `2 + 2.5` | `4.5` |
| `-` | Subtraction | Subtracts values | `10 - 3` | `7` |
| `*` | Multiplication | Multiplies values | `4 * 5` | `20` |
| `/` | Division | Standard division (**always returns a float**) | `5 / 2` | `2.5` |
| `**` | Exponentiation | Power of a number | `2 ** 3` | `8` |
| `//` | Floor Division | Divides and discards decimal (rounds down) | `5 // 2` | `2` |
| `%` | Modulo | Returns the **remainder** of division | `5 % 2` <br> `20 % 7` | `1` <br> `6` |

> 💡 **Key Insight:** Mixing `int` and `float` in operations is valid. The result will usually be a `float` if division is involved.

## 3. Operator Precedence (PEMDAS Rule)
*Goal: Understand the order in which Python solves complex equations.*

Python follows the standard mathematical order of operations:
1.  **P**arentheses `()`
2.  **E**xponents `**`
3.  **M**ultiplication `*`, **D**ivision `/`, **F**loor Division `//`, **M**odulo `%` *(Left to Right)*
4.  **A**ddition `+`, **S**ubtraction `-` *(Left to Right)*

### Walkthrough Examples

**Example 1:** `3 ** 2 + (2 + 3) * 4 - 2`

1.  **Parentheses:** `(2 + 3)` becomes `5`
    *   *Equation:* `3 ** 2 + 5 * 4 - 2`
2.  **Exponents:** `3 ** 2` becomes `9`
    *   *Equation:* `9 + 5 * 4 - 2`
3.  **Multiplication:** `5 * 4` becomes `20`
    *   *Equation:* `9 + 20 - 2`
4.  **Addition/Subtraction:** `9 + 20` is `29`, then `29 - 2`
    *   **Final Result:** `27`

**Example 2:** `25 // 2 + 3 ** 2`

1.  **Exponents:** `3 ** 2` becomes `9` *(Exponents come before Floor Division)*
    *   *Equation:* `25 // 2 + 9`
2.  **Floor Division:** `25 // 2` becomes `12` *(Discards the .5)*
    *   *Equation:* `12 + 9`
3.  **Addition:** `12 + 9`
    *   **Final Result:** `21`

*   **The Code:**

    *Result:* `9125` days.

*   **Adding Complexity (Hours):**
    If we want hours, we multiply by 24.
    `25 * 365 * 24`
    *Result:* `219,000` hours.

> **Lesson:** Multiplication is associative, so order doesn't change the result here, but if we added "leap years" (adding 1 day every 4 years), PEMDAS would ensure we add the extra days *after* calculating the base years.
