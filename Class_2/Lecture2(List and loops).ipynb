# Creating and Defining Lists
* Syntax: Lists are defined using box brackets (square brackets) [].
* Heterogeneous Data: Unlike some other data types, a Python list can hold different types of data simultaneously. The instructor demonstrates creating a list that contains a string (Name), an integer (Age), and a float/integer (Salary) all in one structure (e.g., ["Danish", 21, 50000])

# Type Checking: 
* Using the type() function confirms that these variables belong to <class 'list'>

# Indexing (Accessing Data)
* Positive Indexing: Starts from 0 and moves from left to right. For example, index 0 accesses the first item.
* Negative Indexing: Starts from -1 and moves from right to left. The last item in the list is always -1. 

# Slicing (Accessing Ranges) When multiple values are needed (e.g., Name and Age, but not Salary)
* Syntax: Uses the format [start:end]

# Modifying Lists (Appending and Updating)
* Appending: The .append() method is used to add new data to the end of an existing list.
* Updating: Values can be changed by referencing their index.
* insert(index, element)

# Dynamic Indexing and Deletion
* Finding Index: If you know a value (e.g., 21) but not its position, the .index(value) method returns the index number (e.g., 1). This allows for dynamic updating without manually counting positions.
* Deleting by Index (pop): The .pop(index) method removes an item based on its index number. 
* Deleting by Value (remove): The .remove(value) method deletes an item by specifying the actual data (e.g., "Karachi") rather than its position
