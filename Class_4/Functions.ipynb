# Python Functions Guide

### 1. Functions vs. Methods

*   **Functions** (such as `print()` and `len()`) are general and can be used across various data types like lists, integers, floats, and strings.
*   **Methods**, however, belong to a specific class. For instance, `.append()` is a method that belongs strictly to the list class and can only be used on lists, not on integers or strings.

---

### 2. Function Declaration and Return Statements

A function is created using the `def` keyword, followed by a valid legal function name, parentheses, a colon, and an indented block of code representing its body. Functions utilize the `return` keyword to pass calculated variables or values back to the user upon completion.

**Example:**

```python
def output():
    value = "Hello World"
    return value
```

When `output()` is called, the function returns and outputs `"Hello World"`.

---

### 3. Parameters and Positional Arguments

Parameters are defined inside the function's parentheses during declaration, and arguments are the actual values passed when calling the function. In positional arguments, the values are assigned strictly based on their sequence or position.

**Example:**

```python
def add(num1, num2):
    addition = num1 + num2
    return addition
    
add(15, 12)
```

In this call, `15` is assigned to the first parameter (`num1`) and `12` is assigned to the second parameter (`num2`), returning `27`.

---

### 4. Keyword Arguments

Keyword arguments allow you to explicitly state the parameter's name alongside its value during the function call, which means you do not have to follow the original positional order.

**Example:**

```python
def test(a, b, c):
    return a + b, b - c

test(b=10, c=5, a=15)
```

Even though the order is different from the declaration, Python correctly assigns `10` to `b`, `5` to `c`, and `15` to `a`.

---

### 5. Mixed Arguments

Mixed arguments combine both positional and keyword arguments in a single function call. However, there is a strict rule: **positional arguments must always be written before keyword arguments.**

**Correct Example:**

```python
test(20, c=15, b=25)
```

Here, `20` acts as a positional argument for the first parameter (`a`), while `c` and `b` are provided as keyword arguments.

**Incorrect Example:**

```python
test(b=10, 14, 12)
```

This causes a syntax error because a keyword argument (`b=10`) is placed before positional arguments.

---

### 6. Default Arguments

Default arguments allow you to define an initial value for a parameter at the time of function declaration. If the user does not provide an argument when calling the function, it uses this default value; if an argument is provided, it overrides the default.

**Example:**

```python
def salary_percentage(salary, tax_rate=0.05):
    return salary * tax_rate
```

*   Calling `salary_percentage(20000)` automatically uses the default `0.05` for `tax_rate`, calculating a tax of `1000.0`.
*   Calling `salary_percentage(20000, 0.10)` overrides the default, using `0.10` as the new `tax_rate`. You can also use keyword arguments to override it, such as `salary_percentage(tax_rate=0.20, salary=20000)`.
